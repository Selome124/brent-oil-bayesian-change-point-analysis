```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load Brent oil price data
brent_df = pd.read_csv('../data/raw/BrentOilPrices.csv', parse_dates=['Date'])

# Show first 5 rows
print(brent_df.head())

# Plot the price over time
plt.figure(figsize=(12,6))
plt.plot(brent_df['Date'], brent_df['Price'], label='Brent Oil Price')
plt.title('Brent Oil Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
